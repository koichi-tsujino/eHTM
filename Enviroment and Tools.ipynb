# ターミナルとSFTPクライアント

Linuxをリモートで使うためには、ターミナルソフト(SSHクライアント）とSFTPクライアント必須です。さまざまなソフトがあります。よく使われれるものとしては、puTTY、teraterm、WinSCPなどがありますが、お薦めは、MobaXterm です。SSHとSFTPクライアントが統合されていて、非常に使いやすくなっています。

## MobaXtermのダウンロードとインストール

http://mobaxterm.mobatek.net/download-home-edition.html にアクセスして、「MobaXterm Home Edition v10.2 (Portable edition)」をクリックして、ZIPファイルのダウンロード＆展開を行ってください。「MobaXterm_Personal_10.2.exe」をダブルクリックすると起動します。

## SSHクライアントとしての使い方
メニューバーから、「Sessions>New Session」を開き、「SSH」をクリックしてください。

Remote host, user name, port, private key を指定します。設定が終わったら、「OK」をクリックします。

## Jupyter Notebook
### インストール

Anaconda には、すでにJupyter notebookが含まれています。

はじめに、ipythonを起動してJupyterログイン時のパスワードを設定します。

     $ ipython

     In [1]: from notebook.auth import passwd

     In [2]: passwd()
       Enter password: <パスワードを入力>
       Verify password: <パスワードを再入力>
       Out[2]: 'sha1:<ハッシュ化されたパスワード>'

     In [3]: quit()

     
     config file を生成します。
     $jupyter notebook --generate-config
